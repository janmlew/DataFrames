```python
import pandas as pd
import numpy as np

indeks = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']
cpc = [809, 731, 588, 18, 200, 70, 45]
cntry = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr = [True, False, False, False, True, True, True]

cars = pd.DataFrame({'cars_per_cap':cpc, 'country':cntry, 'drives_right':dr}, index=indeks)

#np.logical_and(cars['cars_per_cap'] > 500, cars['cars_per_cap'] < 800)

print('Countries in which cars drive right lanes:')
sel = cars[cars["drives_right"]]
print(sel)
#sel = cars.loc[:,'drives_right']
#print(cars[sel])

z = 6
if z%2 == 0:
    print('z is an even number.')
elif z%3 == 0:
    print('z is divisible by 3.')
else:
    print('z is an odd number.')

error = 50.0
while error > 1:
    error = error / 4
    print(error)

for index, country in enumerate(cars['country']):
    print('Index = ' + str(index) + ", and country = " + str(country))

print(type(enumerate(cars)))

house = [["hallway", 11.25],
         ["kitchen", 18.0],
         ["living room", 20.0],
         ["bedroom", 10.75],
         ["bathroom", 9.50]]
for sublist in house:
    print('the ' + str(sublist[0]) + ' is ' + str(sublist[1]) + ' sqm')

world = {'afganistan':30.55,
         'albania':2.77,
         'algeria':39.21}
for key, value in world.items():
    print(key + ' -- ' + str(value))

np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
meas = np.array([np_height, np_weight])
for val in np.nditer(meas):
    print(val)

for x, y in cars.iterrows():
    cars.loc[x, 'country_len'] = len(y['country'])
#This for function adds a column named 'country_len'
# with the length of a string in the 'country' column

cars['country_len'] = cars['country'].apply(len)

print('so it begins')
for i, x in cars.iterrows():
    print(i + ': ' + str(x['country_len']))
print('and so it ends')

cars['COUNTRY'] = cars['country'].apply(str.upper)

np.random.seed(123)
print(np.random.rand())
print(np.random.rand())
print(np.random.randint(0,2))
```
